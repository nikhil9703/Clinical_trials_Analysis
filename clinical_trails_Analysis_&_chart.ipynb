```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('clin_trials.csv')
```

```python
data.head()
```

```python
organization_class = data['Organization Class'].value_counts()
print(organization_class)
```

```python
plt.figure()
plt.bar(organization_class.index , organization_class.values)
plt.xticks(rotation = 45 , fontsize = 'small' ) #for x axis names rotations
plt.xlabel('Orgnization class names')
plt.ylabel('Count of organization class')
plt.title('clinical trials by Organization Class')
plt.show()
```

```python
organization_Full_Name = data['Organization Full Name'].value_counts()

organization_Full_Name_head = organization_Full_Name.head(10)

print(organization_Full_Name_head)
```

```python
plt.figure()
plt.bar(organization_Full_Name_head.index , organization_Full_Name_head.values)
plt.xticks(rotation = 90 , fontsize = 'small' ) #for x axis names rotations
plt.xlabel('Orgnization full names list')
plt.ylabel('Count of organization name')
plt.title('clinical trials by Organization Class')
plt.show()
```

```python
overall_status = data['Overall Status'].value_counts()
print(overall_status)
```

```python
plt.figure()

plt.pie(
    overall_status.values,
    autopct='%1.1f%%',
    pctdistance=1.1,
    textprops={'fontsize':8}
)

plt.legend(
    overall_status.index,
    title="Overall Status",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.axis('equal')
plt.title('Overall Status Distribution')
plt.show()
```

```python
plt.figure()
bars = plt.bar(overall_status.index , overall_status.values)
plt.xticks(rotation = 90)
plt.xlabel('Overall Status used')
plt.ylabel('Total')
plt.bar_label(bars, padding=3, fontsize = 6)
plt.show()
```

```python
primary_purpose = data['Primary Purpose'].value_counts()
print(primary_purpose)
```

```python
plt.figure()
bars = plt.bar(primary_purpose.index , primary_purpose.values)
plt.xticks(rotation = 90)
plt.xlabel('primary purpose')
plt.ylabel('Total')
plt.bar_label(bars, padding=3, fontsize = 6)
plt.show()
```

```python
data['Start Date Parsed'] = pd.to_datetime(
    data['Start Date'],
    format='%d-%m-%Y',
    errors='coerce'
)
cleaned_data = data[data['Start Date Parsed'].notna()].copy()
cleaned_data['Start Date'] = cleaned_data['Start Date Parsed']
cleaned_data.drop(columns=['Start Date Parsed'], inplace=True)
cleaned_data.to_csv('clin_trials_cleaned.csv', index=False)
print("Original rows:", data.shape[0])
print("Cleaned rows:", cleaned_data.shape[0])
```

```python
data1 = pd.read_csv('clin_trials_cleaned.csv')
```

```python
data1['Start Date']
```

```python
data1['Start Date'] = pd.to_datetime(
    data1['Start Date'],
    format='%Y-%m-%d',
    errors='coerce'
)
```

```python
year_counts = (
    data1[data1['Start Date'].dt.year <= 2025]['Start Date']
    .dt.year
    .value_counts()
    .sort_index()
)

print(year_counts)

```

```python
plt.figure()
sct = plt.scatter(year_counts.index, year_counts.values)
plt.xticks(rotation = 90)
plt.xticks(range(2000, 2026, 1))  # every year
plt.xlim(2000, 2025)
plt.xlabel('Year')
plt.ylabel('Total Trials')

for x, y in zip(year_counts.index, year_counts.values):
    if 2000 <= x <= 2025:
        plt.text(x, y, str(y), fontsize=6,
                 ha='center', va='bottom')
plt.show()
```
